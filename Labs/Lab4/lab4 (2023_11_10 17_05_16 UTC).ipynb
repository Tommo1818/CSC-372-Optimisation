```python
import scipy.stats as sps
import numpy as np
import matplotlib.pyplot as plt

print(sps.norm.__doc__)
```

task 1

```python
#create a normal distribution with mean 163 and standard deviation 10
norm = sps.norm(loc=163, scale=10)
#generate 10000 samples from this distribution
hsample = norm.rvs(size=10000)
#plot the histogram of the sample
plt.hist(hsample, bins=100, density=True)
plt.title('Normal distrubution of heights')
plt.xlabel('Height (cm)')
plt.ylabel('Density')

plt.show()
```

task 2

```python
mu = 163
sigma = 10
x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 1000),
y = sps.norm.pdf(x, mu, sigma)
plt.scatter(x, y, s=1)
plt.title('Normal distribution of Heights')
plt.xlabel('Height (cm)')
plt.ylabel('Density')
plt.show()  



```
